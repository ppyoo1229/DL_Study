```python
a = [1, 2, 3, 4, 5]
print(a)
print(len(a))
print(a[0])
print(a[4])
a[4] = 99
print(a)
print(a[0:2])
print(a[1:])
print(a[:3])
print(a[:-1])
print(a[:-2])
```

```python
# 넘파이 가져오기
import numpy as np
```

```python
# 넘파이 가져오기
# 넘파이 배열 생성하기
x = np.array([1.0, 2.0, 3.0])
print(x)  # [1. 2. 3.]
print(type(x))  # <class 'numpy.ndarray'>
```

```python
# 넘파이의 산술 연산
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x + y)  # 원소별 덧셈
print(x - y)
print(x * y)  # 원소별 곱셈(element-wise product)
print(x / y)
```

넘파이 배열은 N차원 배열을 작성할 수 있음.
수학에서는 1차원 배열은 벡터, 2차원 배열은 행렬이라고 부른다.
그리고 이를 일반화 한 것을 텐서라 한다.

```python
# 넘파이의 N차원 배열
A = np.array([[1, 2], [3, 4]])
print(A)
```

```python
print(A.shape)  # 행렬의 형상 (2, 2)
print(A.dtype)  # 행렬에 담긴 원소의 자료형 int64
```

```python
B = np.array([[3, 0], [0, 6]])
print(A + B)
print(A * B)
print(A * 10)
```

```python
# 브로드캐스트
# 넘파이에서는 형이 다른 배열끼리도 계산할 수 있음
A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])
print(A * B)
```

```python
# 원소 접근
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
```

```python
print(X[0])  # [51 55] 0행
print(X[0][1])  # 55 (0, 1)
for row in X:
    print(row)
```

```python
X = X.flatten()  # X를 1차원 배열로 변환(평탄화)
print(X)  # [51 55 14 19  0  4]
print(X[np.array([0, 2, 4])])  # 인덱스가 0,2,4인 원소 얻기 [51 14  0]
print(X > 15)  # [ True  True False  True False False]
               # 넘파이 배열에 부등호 연산자를 사용한 결과는 bool 배열
print(X[X > 15])  # [51 55 19]
```

```python
# matplotlib는 그래프를 그려주는 데이터 시각화 라이브러리
# 그래프를 그리려면 pyplot모듈을 이용
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1) # 0~6 0.1 간격으로 생성
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.show()
```

```python
# 데이터 준비
x = np.arange(0, 6, 0.1) 
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```
